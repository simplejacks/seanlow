```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
import matplotlib.pyplot as plt
import os
import glob
```

```python
def random_floats(low, high, size):
    return [np.random.uniform(low, high) for _ in range(size)]

size = 10000
ep = np.random.normal(0, 1, size).reshape(-1,1)


elements = [0, np.log(0.5)]
probabilities = [0.99,0.01]
zeta=np.random.choice(elements, 10, p=probabilities)

log_g = 0.02+0.02*ep+zeta
g = np.exp(log_g)
g_list = g.tolist

xminus = 1
xplus = 1.1
x = 0.5*(xminus+xplus)
R = x*g
b=3

###Error e(x) function###
def error(x,b):
    R = x*g
    nu_list=[]
    
    for i in R:
        if i <1.0303:
            nu = 2*(i-1.0303)        
        else:
            nu = (i-1.0303)           
        nu_list.append(nu)
        
    nu_mean=np.asarray(nu_list).mean()
    e = 0.99*b*nu_mean+0.99*x-1
    return e
###########################
 
###Bisect function###
def bisect(xminus,xplus,b):
    x = (xminus+xplus)/2
    e = error(x,b)
    if np.absolute(e) < 0.0001:
        return(x)
    elif e < 0:
        return(bisect(x,xplus,b))
    elif e > 0:
        return(bisect(xminus,x,b))
###########################

###Getting values of x for 0<b<10 ###
x_list = []
b0_list = []

for b0 in np.arange(0,10,0.1):
    x = bisect(1,1.1,b0)
    x_list.append(x)
    b0_list.append(b0)

x_array = np.asarray(x_list)
x_df = pd.DataFrame(x_array, columns=['x'])
  
###########################

###Price-Dividend Ratio ###

P_D = 1/(x_df-1)
P_D.rename(columns={'x':'P/D'})
PD_list = P_D.values.tolist()

###########################

###Plot price-dividend ratio vs b0###
plt.style.use('seaborn')
plt.scatter(x = b0_list , y = PD_list)
plt.xlabel('b0')
plt.ylabel('P/D')
plt.title('P/D vs b0')
plt.show()

###########################

###Expected Mkt Return###
E_Rm = (x_array*g[:,None,None]).mean(0)
#print(E_Rm)

###Plot the equity premium (x*g) vs b0 ###
Rf = 1.0303
EP = E_Rm-Rf
#print(EP)
plt.style.use('seaborn')
plt.figure(1)
plt.scatter(x = b0_list , y = EP)
plt.ylim(0, 0.006)
plt.xlabel('b0')
plt.ylabel('Equity Premium')
plt.title('Equity Premium vs Scalability Factor (b0)')
plt.show()
```
